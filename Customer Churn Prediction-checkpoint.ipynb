```python
#Grouping tenure
def tenure_bracket(tenure):
    if tenure <=35:
        return"<=25 Youth"
    elif tenure <=55:
        return"Adult(<=55)"
    elif tenure <=65:
        return "Old Adult(>=65)"
    else:
        return"Elders(>=65)"
df["tenure_bracket"]=df['tenure'].apply(tenure_bracket)
plt.figure(figsize=(10,5))
sns.countplot(x='tenure_bracket', data=df)
plt.xlabel('tenure_group')
plt.ylabel('Count of tenure_group')
plt.title('Total number of customer');
```

# Feature engineering/Data preprocessing

```python
df.head()
#the created features which were created above are not needed, hence will be removed for this process
#create a copy of the data frame
```

```python
df.columns
```

```python
df1=df[['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']]
label=df[['Churn']]
```

```python
df1.head()
```

```python
df1['gender'].replace(['Female', 'Male'],
                        [0, 1], inplace=True)
```

```python
df1.head()
```

```python
label.head()
#target which is the answer to the marking script label need s to be removed for predictions to be carried out properly and avoid overfitting
```

```python
df1.dtypes
```

```python
#dealing with outliers through data normalization
Scaler=MinMaxScaler()
df1['Scaled_PS']=Scaler.fit_transform(df1['PhoneService'].values.reshape(-1,1))
df1['Scaled_GN']=Scaler.fit_transform(df1['gender'].values.reshape(-1,1))
df1['Scaled_TN']=Scaler.fit_transform(df1['tenure'].values.reshape(-1,1))
df1['Scaled_DP']=Scaler.fit_transform(df1['Dependents'].values.reshape(-1,1))
df1['Scaled_MC']=Scaler.fit_transform(df1['MonthlyCharges'].values.reshape(-1,1))
df1.drop(['PhoneService','gender','tenure','Dependents','MonthlyCharges'],axis=1, inplace=True)
df1.head()
```

```python
#split the dataset into training and testing sets
x_train, x_test,y_train, y_test=train_test_split(df1, label, test_size=0.2, random_state=42)
#test size indicate the dataset to test and this represents 20%. Therefore 80% will be used for training
```

```python
#Model building using logistic regression
logreg=LogisticRegression()

logreg.fit(x_train, y_train)

ly_pred=logreg.predict(x_test)

print("Logistic Regression")
print("Accuracy:", accuracy_score(y_test, ly_pred))
print("Precision:", precision_score(y_test, ly_pred))
print("Recall:", recall_score(y_test, ly_pred))
print("F1_score:", f1_score(y_test, ly_pred))
print("AUC_ROC:", roc_auc_score(y_test, ly_pred))
```
